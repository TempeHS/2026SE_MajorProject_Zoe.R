## Sprint 2 - Item Interactions
In the basic setup, I coded in some basic features, including detecting if someone has clicked on an object. 

This sprint will keep track of the amount of times a player has clicked on the object, and bring up dialogue when the object is clicked on. 
### Issues
- When using the variable for detecting mouse clicks, these are detected for however long the mouse is clicked for rather than once. While I could possibly implement some kind of limiter for this, the easier solution is to just use the pygame event handler instead. 
- Ran into the issue of having a while loop without an end condition (causing the game to freeze). Eventually, it ended up working with just an if statement instead of while 
- There were a few issues with the logic of the basic dialogue function, but I managed to get it to work fairly quickly. The same cannot be said for the choice function (though I did assume it would be harder, as there were a lot more factors I had to account for)
- Had an issue with removing items from the dialogue list due to the functions for it changing the list indexes, which I fixed by reformatting the file 
- Had to figure out a way to get the function to skip through lines which weren't relevant to the specific choice that was picked (which was the part that took the longest), as my initial method of just using pass would bring up a blank dialogue box (since the player still needed to press space to progress, and the index of the dialogue brought up wasn't at the correct choice). I also ran into this issue again because I forgot to add the function which updates the display.
- Throughout attempting to program this, the most consistent issue I ran into was out of bounds exceptions, which I had to account for in many areas that I didn't with the dialogue function (since the choice function messed around with which index was called, whereas the dialogue function was consistent with it)
- Additionally, I had to use a few global variables, as simply instantiating them within the function didn't let me change the value inside the function and have that value remain changed (and regardless of how I went about this, I would likely have had to make some variables global)


## Sprint 3 - Save System
This sprint will create a save system, which keeps track of any variables which may change, such that the player can save these and reload them at any point. Loading a save without saving anything will load the last save from previous gameplay, meaning players can close the game and pick up where they left off. 

Currently this system is very simple, and only keeps track of the dialogue interactions a player has had. 

### Issues
- I had very little issues with the general SQL, and creating a system to load the save files
- I had a couple of small issues with integrating it into the dialogue system, with the biggest being to do with how I had the choice system set up (where it would save the choice values into a list, so that I could include dialogue before them). The issue with this was that it would keep adding these values to the list, causing duplicates to display. I ran through a few methods to fix this, but settled on a function that only kept unique values in the list.
